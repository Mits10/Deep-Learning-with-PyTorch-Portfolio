Creating a simple linear layer neural network with torch package

```python
import torch.nn as nn
input_tensor = torch.tensor([[0.3471,0.4547, -0.2356]])
linear_layer = nn.Linear(3,2)
output = linear_layer(input_tensor)
print(output)
print(linear_layer.weight)
print(linear_layer.bias)
```

a simple linear layer without torch package

```python
input_tensor = torch.tensor(3. , requires_grad = True)
output = input_tensor**2
output.backward()
print(input_tensor.grad)
```

We can stack linear layer with nn.Sequential()

```python
model = nn.Sequential(nn.Linear(8,4),
                      nn.Linear(4,2),
                      nn.Sigmoid())
total = 0
for param in model.parameters():
    total += param.numel()
print(total)
```
